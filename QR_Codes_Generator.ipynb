```python
""" 
 Please make sure you have python 3.9 installed with pip privilage on your machine...
 Please make sure the following libraries are installed and are working...
 pyqrcode... to install it use "pip install pyqrcode"
 PIL... to install it use  the following...
 FOR WINDOWS...
     py3 -m pip install --upgrade pip
     py3 -m pip install --upgrade Pillow
     
 FOR MACOS AND LINUX...
     python3 -m pip install --upgrade pip
     python3 -m pip install --upgrade Pillow
     
     
 The datetime and random libraries are already installed with python.
 """


import pyqrcode
from PIL import Image, ImageDraw, ImageFont
import random
from datetime import datetime
```

```python
"""
 The following code generates random codes for dummy vouchers ...
 in order to test the generation and then validate the required QR codes. 
 
 In the first round the code generates only 10 codes, please change the value of 10 in the 'for' loop...
 to generate more code as you like.
"""


codes =[]

for i in range(10):    
    area = random.randint(1, 5) #This generates random integers between 1 and 5 excluding 5.
    ngo = random.randint(1,7) # This generates random integers between 1 and 7 excluding 7.
    ben = random.randint(0,5001) # This generates random integers between 1 and 5001 excluding 5001.
    month = random.randint(1, 13)
    year = random.randint(22, 33)
    med = random.randint(0, 10000)
    proj = random.randint(0, 100)

    code = str("{:02.0f}".format(area))+'-'+str("{:02.0f}".format(ngo))+'-'+str("{:04.0f}".format(ben))+'-'+str("{:02.0f}".format(month))+'-'+str("{:02.0f}".format(year))+'-'+str("{:04.0f}".format(med))+'-'+str("{:02.0f}".format(proj))
    codes.append(code)
    
codes
```

```python
"""
    The following function generates a QR code out of a code generated by the previous code
    To add a logo to the QR code, please plug in the appropriate directory of the logo you would...
    like to add to the QR code at the variable 'logo'.
    
    To modify the font, please add the appropriate directory of the font you would like...
    to use at the object 'font'.
    
    Finally, to save a QR code, please add the directory to which you would...
    like to save your QR code according to your system.

"""

def GenerateQRCode(string_code):
    QR = pyqrcode.QRCode(string_code, error = 'H')
    QR.png('MyQR.png', scale=10)
    im = Image.open('MyQR.png')
    im = im.convert("RGBA")
    logo = Image.open('/home/abdul/Pictures/6347023_preview.png')
    box = (175,175,275,275)
    im.crop(box)
    region = logo
    region = region.resize((box[2] - box[0], box[3] - box[1]))
    im.paste(region,box)
    font = ImageFont.truetype("/usr/share/fonts/truetype/noto/NotoSansDisplay-Regular.ttf", size=15)
    ImageDraw.Draw(im).text((165,0), 'some ID number', (0,0,0), font=font)
    im.save('/home/abdul/Pictures/QR_Codes/{}'.format(string_code), 'png')


```

```python
"""
    Finally, Run this code to generate multiple QR codes according to the random generated dummy...
    voucher codes. The output of this code will show you the execution time according to the...
    specifications of your machine and the length of the codes list 
"""

start_time = datetime.now()
for c in codes:
    GenerateQRCode(c)
    
end_time = datetime.now()
print(end_time - start_time)

```
